```python
## with the help of count plot we will visulaize the canceltion and not cancelation in both hotels 

plt.figure(figsize=(8, 4))
ax = sns.countplot(x='hotel', hue='is_canceled', data=df, palette='Blues')
legend_labels, _ = ax.get_legend_handles_labels() ## handles represnt the grpahicalelements(bar) and labels categories
ax.legend(legend_labels, ['Not Canceled', 'Canceled'], bbox_to_anchor=(1, 1)) ## bbox_to_anchor=(1, 1)) -represnt postion
plt.title('Reservation Status in Different Hotels', size=20) ## represnt titles with the front size 
plt.xlabel('Hotel')
plt.ylabel('Number of Reservations')
plt.show()
```

```python
## we can calculate individual canceliation or not cancelition 
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel = df[df['hotel'] == 'Resort Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
#### we will check the price effect of hotel canceliation 
## we will use mean of Adr ( average daily rate ) with the help of group by 
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel =   city_hotel.groupby('reservation_status_date')[['adr']].mean()

plt.figure(figsize=(20, 8))
plt.title('Average daily rate of city and resort hotel', fontsize=30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label='Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label='City Hotel')
plt.legend(fontsize=15)
plt.show()

```

## output 

The line graph above shows that, on certain days the average daily rate of city hotel is less than that of resort hotel, and on other days, itâ€™s even less. Its goes without saying that weekdays and holidays may see a rise in resort hotel rates.

```python
### We will create a count bar plot to analyze the monthly reservations and cancellations.


df['month'] = df['reservation_status_date']. dt.month
plt.figure(figsize=(16, 8))
ax1 = sns.countplot(x='month', hue='is_canceled', data=df, palette='bright')
legend_labels, _ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1, 1))
plt.title('Reservation Status per Month', size=20)
plt.xlabel('Month')
plt.ylabel('Number of Reservations')
plt.legend(['Not Canceled', 'Canceled'])
plt.show()
```

## output 

We have developed the grouped bar graph to analyses the months with the highest and lowest reservations level according to reservations status. As can be seen, both the number of canceled reservations are largest in the month of August. whereas January is the month with the most canceled reservations.

```python
## in above visulaize we can see in jan month the cancelattion is very high so with the help of barplot 
# we will check whether due to price the cancelation is affected or not 


import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 8))
plt.title('ADR for Each Month', fontsize=30)
sns.barplot(x='month', y='adr', data=df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())
plt.xlabel('Month')
plt.ylabel('Average Daily Rate (ADR)')
plt.show()
```

## output 
The bar graph demonstrates that cancelations are most common that when prices are greatest and are least common that when they are lowest. Therefore, the cost of the accommodation is solely responsible for the cancelation.

```python
## we will check the price of hotel  whether its cancelation or not cancelation 
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

not_cancelled_data = df[df['is_canceled'] == 0]

not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

plt.figure(figsize=(20, 6))
plt.title('Average Daily Rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='not_cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='cancelled')
plt.legend()
plt.show()
```

```python
## ## in above visual data is not consistent to we will filter the data 2016 to 2017-09

cancelled_df_adr = cancelled_data[cancelled_data['reservation_status_date'].between('2016', '2017-09')]
cancelled_df_adr = cancelled_df_adr.groupby('reservation_status_date')[['adr']].mean().reset_index()

not_cancelled_df_adr = not_cancelled_data[not_cancelled_data['reservation_status_date'].between('2016', '2017-09')]
not_cancelled_df_adr = not_cancelled_df_adr.groupby('reservation_status_date')[['adr']].mean().reset_index()

plt.figure(figsize=(20, 6))
plt.title('Average Daily Rate',fontsize =30)
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='not_cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='cancelled')
plt.legend()
plt.show()
```
